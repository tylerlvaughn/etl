## Read csv file

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "combined.csv"

# Read the File and store into Pandas data frame
combined_data = pd.read_csv(file_to_load)
combined_data
```

## Clean the data

```python
#Remove NAN values
combined_data.dropna(how='any', inplace=True)
#Update some column names
combined_data.rename(columns={'urbanization_rate_country':'Country'}, inplace=True)
combined_data.rename(columns={'ladder_score':'happiness_score'}, inplace=True)
#show df
combined_data
```

```python
#round values in some columns
combined_data['happiness_score'] = (combined_data['happiness_score'].round(2))
combined_data['logged_gdp_per_capita'] = (combined_data['logged_gdp_per_capita'].round(2))
combined_data['social_support'] = (combined_data['social_support'].round(2))
combined_data['healthy_life_expectancy'] = (combined_data['healthy_life_expectancy'].round(2))
combined_data['freedom_to_make_life_choices'] = (combined_data['freedom_to_make_life_choices'].round(2))
combined_data['generosity'] = (combined_data['generosity'].round(2))
combined_data['perceptions_of_corruption'] = (combined_data['perceptions_of_corruption'].round(2))
combined_data
```

## Load Data into PyMongo

```python
#import dependencies
import pymongo

# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define database and collection
combined_db = client.combined_db
combined_collection = combined_db.items

#covert dataframe to Dictionary
dict = combined_data.to_dict("list")
print(dict)
```

```python
# Load Collection 
for i in range(len(dict['Country'])):
    post = {
        'Country': dict['Country'][i],
        'urbanization_rate': dict['urbanization_rate'][i],
        'suicide_rate': dict['suicide_rate'][i],
        'meat_consumption': dict['meat_consumption'][i],
        'gdp_per_capita': dict['gdp_per_capita'][i],
        'fertility': dict['fertility'][i],
        'happiness_score': dict['happiness_score'][i],
        'logged_gdp_per_capita': dict['logged_gdp_per_capita'][i],
        'social_support': dict['social_support'][i],
        'healthy_life_expectancy': dict['healthy_life_expectancy'][i],
        'freedom_to_make_life_choices': dict['freedom_to_make_life_choices'][i],
        'generosity': dict['generosity'][i],
        'perceptions_of_corruption': dict['perceptions_of_corruption'][i]
         }
    print(post)
    combined_collection.insert_one(post)
```

```python
#query database to confrm load was sucessful
happiness_scores = combined_db.items.find()
for item in happiness_scores:
    print(item)
```
